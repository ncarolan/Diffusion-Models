## Attention-Model

#### A project by Nicholas Carolan, Caiti Hensley, Wayne Nyabuto

This network spreading model considers the role of attention span in modeling the spread of memes on social networks. Agents maintain a limited attention span (queue) of their most recently observed memes. They draw from this queue to re-share memes.

```python
from queue import Queue 
import networkx as nx
import collections
import random
import matplotlib.pyplot as plt
```

```python
class Agent:
    
    def __init__(self,ID,attention_span,interests):
        self.ID = ID
        self.attention_span = attention_span
        self.dq = collections.deque(maxlen=attention_span)
        self.interests = interests
        self.meme_share_chance = abs(random.random() - random.random()) #weighted towards 0
        self.meme_create_chance = abs(random.random() - random.random()) #weighted towards 0
        
    def add_meme(self,meme):
        self.dq.append(meme)
        
    def select_meme(self):
        if(len(self.dq) == 0):
            return None
        else:
            return random.choice(list(self.dq)) #Randomly select meme from queue
```

```python
class Meme:
    
    def __init__(self,ID,quality,topic):
        self.ID = ID
        self.quality = quality
        self.topic = topic 
```

## Initialize Network

We intialize a network from a sample of 81,000 accounts on Twitter, taken from the Stanford Large Network Dataset Collection (http://snap.stanford.edu/data/ego-Twitter.html). For each account in the network, we model an attention span with a FIFO queue with size randomly selected between 5-10 memes. Each account has a random chance to create memes and a random chance to share memes, both weighted toward zero.

```python
g = nx.DiGraph()
```

```python
import gzip
import csv

following = {}

with gzip.open('twitter_combined.txt.gz', "rt", encoding="utf-8") as input_file:
    reader = csv.reader(input_file, delimiter=' ', quotechar='"')
    for record in reader:
        if(int(record[0]) not in following.keys()):
            following[int(record[0])] = [int(record[1])]
        else:
            following[int(record[0])].append(int(record[1]))
```

```python
for node in following.keys():
    g.add_node(node)
for node in following.keys():
    for target in following[node]:
        g.add_edge(target,node) #Edge Direction from followee to follower - edges point outward toward followers
```

### Initialize Agents

```python
agents = {}
for node in g.nodes():
    agent = Agent(node, random.randint(5,10), [])
    agents[node] = agent
```

### Initiate Meme Spread

```python
meme_id_counter = 0
```

```python
def initiate_meme_spread(agents):
    global meme_id_counter
    for agent in agents.keys():
        if random.random() < agents[agent].meme_create_chance:
            new_meme = Meme(meme_id_counter, random.randint(0,10), random.randint(0,20))
            agents[agent].add_meme(new_meme)
            meme_id_counter = meme_id_counter + 1
    return agents
```

### Iterate Meme Spread

```python
def iterate_meme_spread(g, agents):
    global meme_id_counter
    for node in g.nodes():
        if random.random() < agents[node].meme_create_chance:
            new_meme = Meme(meme_id_counter, random.randint(0,10), random.randint(0,20))
            agents[node].add_meme(new_meme)
            meme_id_counter = meme_id_counter + 1
            #for neighbor in g.neighbors(node):
            #        agents[neighbor].add_meme(new_meme)
        if random.random() < agents[node].meme_share_chance:
            meme = agents[node].select_meme()
            if not meme == None:
                for neighbor in g.neighbors(node):
                    agents[neighbor].add_meme(meme)
    return agents
```

### Analyze Meme Spread

```python
def analyze_meme_spread(agents):
    memes = {}
    for agent in agents.keys():
        for meme in agents[agent].dq:
            if not meme.ID in memes.keys():
                memes[meme.ID] = 1
            else:
                memes[meme.ID] = memes[meme.ID] + 1
    sorted_memes = {k: v for k, v in sorted(memes.items(), key=lambda item: item[1], reverse=True)}
    return sorted_memes
```

## Simulate Meme Spread

We simulate the creation and spread of unique memes (referenced by their IDs) over our Twitter network. During each iteration, accounts have a chance to create a new meme and/or share an existing meme. In both instances, they share the considered meme with all other accounts that follow them. Those followers can consider the shared meme while it is still in their attention span.

```python
def simulate_meme_spread(g, agents, iterations):
    agents = initiate_meme_spread(agents)
    meme_analysis_series = []
    for x in range(0, iterations):
        agents = iterate_meme_spread(g, agents)
        meme_analysis = analyze_meme_spread(agents)
        print({k: meme_analysis[k] for k in list(meme_analysis)[:5]})
        meme_analysis_series.append(meme_analysis)
    return agents, meme_analysis_series
```

The list below shows the most popular memes during each iteration of meme spread, formatted as  *meme_id: count*

```python
spread,time_series = simulate_meme_spread(g, agents, 10)
```

## Visualize Meme Spread

The visualizations below plot the number of instances of the top 5 memes during each iteration over the entire duration of meme spread.

```python
def plot_meme_spread(iterations, time_series, top_range):
    top_memes = {}
    X = range(iterations)
    
    #Populate top memes
    for i in range(iterations):
        labels = (list(time_series[i].keys())[:top_range])
        for label in labels: 
            if label not in top_memes.keys(): top_memes[label] = []
    
    for i in range(iterations):
        for meme in top_memes.keys(): 
            if(meme not in time_series[i].keys()): 
                top_memes[meme].append(0)
            else: 
                top_memes[meme].append(time_series[i][meme])
        
    #Graph memes over time
    for meme in top_memes: 
        plt.title('Top Memes over Time')
        plt.xlabel('Iterations')
        plt.ylabel('Intances of Meme')
        plt.plot(X,top_memes[meme],label=meme)
```

The visualization below shows the trends in top memes for agents with attention spans between 5 and 10 memes.

```python
plot_meme_spread(10,time_series,5)
```

### Impatient Agents

We consider *impatient* agents to be accounts with very short attention spans. Here we simulate meme spread among accounts with attention spans of 2-5 memes.

```python
impatient_agents = {}
for node in g.nodes():
    agent = Agent(node, random.randint(2,5), [])
    impatient_agents[node] = agent
```

```python
imp_spread,imp_time_series = simulate_meme_spread(g, impatient_agents, 10)
```

```python
plot_meme_spread(10,imp_time_series,5)
```

### Patient Agents

We consider *patient* agents to be accounts with long attention spans. Here we simulate meme spread among accounts with attention spans between 10-25 memes.

```python
patient_agents = {}
for node in g.nodes():
    agent = Agent(node, random.randint(10,25), [])
    patient_agents[node] = agent
```

```python
pat_spread,pat_time_series = simulate_meme_spread(g, patient_agents, 10)
```

```python
plot_meme_spread(10,pat_time_series,5)
```

These patient agents produced markedly different trends in the top memes. In networks of less patient agents, the most popular memes emerged, changed, and declined frequently. However, in this simulation of patient agents, the top memes were remarkably consistent over all iterations. Additionally, the most popular memes in this simulation had considerably more instances - they occupied a greater 'market share' of the total volume of memes. In this patient scenario, a few memes came to dominate the network and endure over time. This is a strikingly different result to the previous simulations.
